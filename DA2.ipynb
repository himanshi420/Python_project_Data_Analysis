```python
import numpy as np
import seaborn as sb
import pandas as pd
import matplotlib.pyplot as plt


%matplotlib inline

```

```python
df= pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')
```

```python
df.head()
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1, inplace=True)
```

```python
# EXPLORATORY DATA ANALYSIS


#Gender

sb.countplot(x ='Gender' , data=df)
```

```python
ax=sb.countplot(x ='Gender' , data=df)

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by ='Amount', ascending = False)
sb.barplot(x= 'Gender',y= 'Amount',data= sales_gen)
```

**From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men**

```python
#AGE

ax = sb.countplot(data = df, x = 'Age Group' , hue = 'Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
#total amount Vs age group

sales_gen = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by ='Amount', ascending = False)

sb.barplot(x='Age Group',y='Amount', data = sales_gen)
```

**From above graphs we can see that most of the buyers are of age group between 26-35 yrs female**

```python
df.columns
```

```python
#state
```

```python
#total number of  order from top 10 states

sales_gen = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by ='Orders', ascending = False).head(10)

sb.set(rc={'figure.figsize':(18,5)})
sb.barplot(data=sales_gen,x= 'State' , y='Orders')

```

```python
#total number of states wise amount top 10

sales_gen = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by ='Amount', ascending = False).head(10)

sb.set(rc={'figure.figsize':(18,5)})
sb.barplot(data=sales_gen,x= 'State' , y='Amount')
```

**From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively**

```python
#marital status

ax= sb.countplot(data= df, x='Marital_Status')

sb.set(rc={'figure.figsize':(8,5)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by ='Amount', ascending = False).head(10)

sb.set(rc={'figure.figsize':(18,5)})
sb.barplot(data=sales_gen,x= 'Marital_Status' , y='Amount', hue='Gender')
```

**From above graphs we can see that most of the buyers are married (women) and they have high purchasing power**

```python
#occupation


sb.set(rc={'figure.figsize':(20,5)})
ax= sb.countplot(data=df, x='Occupation')

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_occp = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by = 'Amount',ascending=False).head(10)

sb.barplot(data= sales_occp, x= 'Occupation', y= 'Amount')


```

**From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector**

```python
df.columns
```

```python
#product category

sb.set(rc={'figure.figsize':(30,8)})
ax= sb.countplot(data=df, x= 'Product_Category')

for ruzz in ax.containers:
  ax.bar_label(ruzz)
```

```python
sales_gen = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by ='Amount', ascending = False).head(10)

sb.set(rc={'figure.figsize':(20,5)})
sb.barplot(data=sales_gen,x= 'Product_Category' , y='Amount')
```

**From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category**

```python
# product id base most order items

sales_gen = df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by ='Orders', ascending = False).head(10)

sb.set(rc={'figure.figsize':(20,5)})
sb.barplot(data=sales_gen,x= 'Product_ID' , y='Orders')
```

**conclusion**

Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category
